### Automatic File Sorter In File Explorer

```python
import os, shutil
```

```python
#change this path to the folder you want to organize
path = r"YOUR_FOLDER_PATH_HERE"
```

```python
file_name = os.listdir(path)
```

```python
folder_names = ['csv files', 'image files', 'text files', 'pdf files', 'docx files', 'vid files']

for loop in range(0,6):
    if not os.path.exists(path + folder_names[loop]):
        print(path + folder_names[loop])
        os.makedirs(path + folder_names[loop])
```

```python
for file in file_name:
    if ".csv" in file and not os.path.exists(path + "csv files/" + file):
        shutil.move(path + file, path + "csv files/" + file)
    elif ".jpg" in file and not os.path.exists(path + "image files/" + file):
        shutil.move(path + file, path + "image files/" + file)
    elif ".TXT" in file and not os.path.exists(path + "text files/" + file):
        shutil.move(path + file, path + "text files/" + file)
    elif ".pdf" in file and not os.path.exists(path + "pdf files/" + file):
        shutil.move(path + file, path + "pdf files/" + file)
    elif ".docx" in file and not os.path.exists(path + "docx files/" + file):
        shutil.move(path + file, path + "docx files/" + file)
    elif ".mp4" in file and not os.path.exists(path + "vid files/" + file):
        shutil.move(path + file, path + "vid files/" + file)
```
